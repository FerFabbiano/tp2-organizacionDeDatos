#Feature Engineering

```
import pandas as pd
import numpy as np
from google.colab import files
```

Cargamos los datasets originales

```
test_set = pd.read_csv('https://raw.githubusercontent.com/FerFabbiano/tp2-organizacionDeDatos/main/Test_TP2_Datos_2020-2C.csv')
train_set = pd.read_csv('https://raw.githubusercontent.com/FerFabbiano/tp2-organizacionDeDatos/main/Train_TP2_Datos_2020-2C.csv')
```

#Total Taxable Amount en dolares

realizamos la conversion del total taxable amount a dolares.

Primero para el trainSet

```
ASP_Promedio_Por_Moneda_train = train_set.groupby('ASP_Currency')['ASP'].transform('mean')
ASP_Converted_Promedio_Por_Moneda_train = train_set.groupby('ASP_Currency')['ASP_(converted)'].transform('mean')

ASP_Es_Nulo_train = train_set.ASP.isnull()
nuevo_ASP_train = ASP_Promedio_Por_Moneda_train if (lambda x: x.ASP_Es_Nulo_train) else train_set.ASP

ASP_Converted_Es_Nulo_train = train_set['ASP_(converted)'].isnull()
nuevo_ASP_Converted_train = ASP_Converted_Promedio_Por_Moneda_train if (lambda x: x.ASP_Es_Nulo_train) else train_set['ASP_(converted)']
```

```
train_set_features1 = pd.DataFrame()
train_set_features1['Total_Amount_USD'] = (nuevo_ASP_Converted_train / nuevo_ASP_train) * (train_set['Total_Amount'])
train_set_features1['Total_Taxable_Amount_USD'] = (nuevo_ASP_Converted_train / nuevo_ASP_train) * (train_set['Total_Taxable_Amount'])
train_set_features1['ID'] = train_set['ID']
```

Ahora para el testSet

```
ASP_Promedio_Por_Moneda_test = test_set.groupby('ASP_Currency')['ASP'].transform('mean')
ASP_Converted_Promedio_Por_Moneda_test = test_set.groupby('ASP_Currency')['ASP_(converted)'].transform('mean')

ASP_Es_Nulo_test = test_set.ASP.isnull()
nuevo_ASP_test = ASP_Promedio_Por_Moneda_test if (lambda x: x.ASP_Es_Nulo_test) else test_set.ASP

ASP_Converted_Es_Nulo_test = test_set['ASP_(converted)'].isnull()
nuevo_ASP_Converted_test = ASP_Converted_Promedio_Por_Moneda_test if (lambda x: x.ASP_Es_Nulo_test) else test_set['ASP_(converted)']
```

```
test_set_features1 = pd.DataFrame()
test_set_features1['Total_Amount_USD'] = (nuevo_ASP_Converted_test / nuevo_ASP_test) * (test_set['Total_Amount'])
test_set_features1['Total_Taxable_Amount_USD'] = (nuevo_ASP_Converted_test / nuevo_ASP_test) * (test_set['Total_Taxable_Amount'])
test_set_features1['ID'] = test_set['ID']
```

# Diferencia entre la creación de la cuenta y la creación de la oportunidad

```
train_set_features2 = pd.DataFrame()

train_set_features2['Account_Created_Date'] = pd.to_datetime(train_set['Account_Created_Date']) 
train_set_features2['Opportunity_Created_Date'] = pd.to_datetime(train_set['Opportunity_Created_Date']) 
train_set_features2['Daily_Dif_Account_Created_Opportunity_Created_Date'] = (train_set_features2['Opportunity_Created_Date'] - train_set_features2['Account_Created_Date']).dt.days
train_set_features2['ID'] = train_set['ID']
```

```
test_set_features2 = pd.DataFrame()

test_set_features2['Account_Created_Date'] = pd.to_datetime(train_set['Account_Created_Date']) 
test_set_features2['Opportunity_Created_Date'] = pd.to_datetime(train_set['Opportunity_Created_Date']) 
test_set_features2['Daily_Dif_Account_Created_Opportunity_Created_Date'] = (test_set_features2['Opportunity_Created_Date'] - test_set_features2['Account_Created_Date']).dt.days
test_set_features2['ID'] = test_set['ID']
```

# Diferencia entre el día que se planea empezar con la entrega, y el día que se planea terminar con la entrega. 

```
train_set_features3= pd.DataFrame()

train_set_features3['Planned_Delivery_Start_Date'] = pd.to_datetime(train_set['Planned_Delivery_Start_Date']) 
train_set_features3['Planned_Delivery_End_Date'] = pd.to_datetime(train_set['Planned_Delivery_End_Date']) 
train_set_features3['Daily_Dif_Delivery_Start_Delivery_End_Date'] = (train_set_features3['Planned_Delivery_End_Date'] - train_set_features3['Planned_Delivery_Start_Date']).dt.days
train_set_features3['ID'] = train_set['ID']
train_set_features3
```

```
test_set_features3= pd.DataFrame()

test_set_features3['Planned_Delivery_Start_Date'] = pd.to_datetime(test_set['Planned_Delivery_Start_Date']) 
test_set_features3['Planned_Delivery_End_Date'] = pd.to_datetime(test_set['Planned_Delivery_End_Date']) 
test_set_features3['Daily_Dif_Delivery_Start_Delivery_End_Date'] = (test_set_features3['Planned_Delivery_End_Date'] - test_set_features3['Planned_Delivery_Start_Date']).dt.days
test_set_features3['ID'] = test_set['ID']
```

# ¿El vendedor dueño de la oportunidad, es el mismo que realizó la última modificación en la misma?

```
train_set_features4= pd.DataFrame()

train_set_features4['Opp_Owner_Did_Last_Modifiy'] = (train_set['Opportunity_Owner'] == train_set['Last_Modified_By'])
train_set_features4['ID'] = train_set['ID']
```

```
test_set_features4= pd.DataFrame()

test_set_features4['Opp_Owner_Did_Last_Modifiy'] = (test_set['Opportunity_Owner'] == test_set['Last_Modified_By'])
test_set_features4['ID'] = test_set['ID']
```

#Cantidad total de trf y de productos por oportunidad

```
train_set_features5 = pd.DataFrame()
train_set['Products_Per_Op'] = 1
train_set_features5['Products_Per_Op'] = train_set.groupby('Opportunity_ID')['Products_Per_Op'].transform('sum')
train_set_features5['TRF_Per_Op'] = train_set.groupby('Opportunity_ID')['TRF'].transform('sum')
train_set_features5['ID'] = train_set['ID']
```

```
test_set_features5 = pd.DataFrame()
test_set['Products_Per_Op'] = 1
test_set_features5['Products_Per_Op'] = test_set.groupby('Opportunity_ID')['Products_Per_Op'].transform('sum')
test_set_features5['TRF_Per_Op'] = test_set.groupby('Opportunity_ID')['TRF'].transform('sum')
test_set_features5['ID'] = test_set['ID']
```

Una vez verificado podemos continuar

```
train_set_features6 = pd.DataFrame()
train_set_features6['Op_Has_A_Valid_Quote_Expiry_Date'] =  (train_set['Quote_Expiry_Date'] != 'NaT')
train_set_features6['ID'] = train_set['ID']
```

```
test_set_features6 = pd.DataFrame()
test_set_features6['Op_Has_A_Valid_Quote_Expiry_Date'] =  (test_set['Quote_Expiry_Date'] != 'NaT')
test_set_features6['ID'] = test_set['ID']
```

# Verificamos si Last_Modified_Date es mayor a Planned_Delivery_End_Date para cada oportunidad

```
train_set_features7 = pd.DataFrame()
train_set_features7['Last_Modified_date>Planned_Delivery_End_Date'] = (pd.to_datetime(train_set['Last_Modified_Date']) > pd.to_datetime(train_set['Planned_Delivery_End_Date']))
train_set_features7['ID'] = train_set['ID']
```

```
test_set_features7 = pd.DataFrame()
test_set_features7['Last_Modified_date>Planned_Delivery_End_Date'] = (pd.to_datetime(test_set['Last_Modified_Date']) > pd.to_datetime(test_set['Planned_Delivery_End_Date']))
test_set_features7['ID'] = test_set['ID']
```

# Producto más vendido por oportunidad

```
#NO ESTA TERMINADO
train_set_features8 = pd.DataFrame()
train_set_features8 = train_set.loc[:,['Opportunity_ID','Product_Name']] 
train_set_features8['Item_Counter'] = 1
train_set_features8['Product_Amount'] = train_set_features8.groupby(['Opportunity_ID','Product_Name'])['Item_Counter'].transform('sum')
train_set_features8['Op_Highest_Amount_Prod(min = 2)'] = 'None'
train_set_features8
#train_set_features8.groupby(['Opportunity_ID','Product_Name'])['Product_Amount'].transform('max')
#train_set_features8.groupby(['Opportunity_ID','Product_Name']).agg({'Product_Amount':'max'})
```

# Unificamos data sets

Primero creamos el train set con los nuevos features

```
train_set_features = pd.DataFrame()

train_set_features = train_set_features2[['Daily_Dif_Account_Created_Opportunity_Created_Date', 'ID']].merge(train_set_features3[['Daily_Dif_Delivery_Start_Delivery_End_Date', 'ID']], on="ID")
train_set_features = train_set_features.merge(train_set_features1[['Total_Amount_USD','Total_Taxable_Amount_USD', 'ID']], on="ID")
train_set_features = train_set_features.merge(train_set_features4[['Opp_Owner_Did_Last_Modifiy', 'ID']], on="ID")
train_set_features = train_set_features.merge(train_set_features5[['Products_Per_Op','TRF_Per_Op', 'ID']], on="ID")
train_set_features = train_set_features.merge(train_set_features6[['Op_Has_A_Valid_Quote_Expiry_Date', 'ID']], on="ID")
train_set_features = train_set_features.merge(train_set_features7[['Last_Modified_date>Planned_Delivery_End_Date', 'ID']], on="ID")
```

```
train_set.drop(columns='Products_Per_Op', inplace=True)

train_set_features = train_set_features.merge(train_set, on="ID")
```

Ahora creamos el test set con los nuevos features

```
test_set_features = pd.DataFrame()

test_set_features = test_set_features2[['Daily_Dif_Account_Created_Opportunity_Created_Date', 'ID']].merge(test_set_features3[['Daily_Dif_Delivery_Start_Delivery_End_Date', 'ID']], on="ID")
test_set_features = test_set_features.merge(test_set_features1[['Total_Amount_USD','Total_Taxable_Amount_USD', 'ID']], on="ID")
test_set_features = test_set_features.merge(test_set_features4[['Opp_Owner_Did_Last_Modifiy', 'ID']], on="ID")
test_set_features = test_set_features.merge(test_set_features5[['Products_Per_Op','TRF_Per_Op', 'ID']], on="ID")
test_set_features = test_set_features.merge(test_set_features6[['Op_Has_A_Valid_Quote_Expiry_Date', 'ID']], on="ID")
test_set_features = test_set_features.merge(test_set_features7[['Last_Modified_date>Planned_Delivery_End_Date', 'ID']], on="ID")
```

```
test_set.drop(columns='Products_Per_Op', inplace=True)

test_set_features = test_set_features.merge(test_set, on="ID")
```
